```python
neverMarried = pres[pres.fmarno ==0]  

```

```python
married = pres[pres.fmarno > 0] 
```

```python
married1 = thinkstats2.Hist(married.totincr, label = 'Married Income')
neverMarried1 = thinkstats2.Hist(neverMarried.totincr, label = 'Never Married Income')
```

```python
width = 0.45
thinkplot.PrePlot(2)
thinkplot.Hist(married1, align = 'right' , width = width)
thinkplot.Hist(neverMarried1, align = 'left', width = width)
thinkplot.Config(xlabel = 'Married and neverMarried Income', ylabel = 'Total Incomes')
```

```python
max_married = married.totincr.max()
max_married
```

```python
min_married = married.totincr.min()
```
